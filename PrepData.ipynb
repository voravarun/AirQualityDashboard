# Load File

```python
import pandas as pd
```

```python
df = pd.read_csv('GlobalWeatherRepository.csv')
df.head()
```

```python
df.shape
```

# Check for duplicate and missing values

```python
# check duplicates
df.duplicated().sum()
```

```python
# check missing values
df.isnull().sum()
```

# Filter for Bangkok

```python
df_bangkok = df[df['location_name'] == 'Bangkok'].copy()
df_bangkok.head()
```

```python
df_bangkok.shape
```

```python
df_bangkok.info()
```

# Convert Datetime datatype

```python
df_bangkok['last_updated'] = pd.to_datetime(df_bangkok['last_updated'])
df_bangkok.info()
```

# Set index
Useful for plotting time series data

```python
df_bangkok.set_index('last_updated', inplace=True)
```

# Feature Engineering

Extract new columns for easier filtering

```python
df_bangkok['month'] = df_bangkok.index.month
df_bangkok['day'] = df_bangkok.index.day
df_bangkok['year'] = df_bangkok.index.year
df_bangkok['hour_of_day'] = df_bangkok.index.time
df_bangkok['day_of_week'] = df_bangkok.index.day_name()
df_bangkok.head()
```

```python
df_bangkok[['month', 'day', 'year', 'hour_of_day', 'day_of_week']]
```

# Handle datatype
Ensure numeric columns data

```python
numeric_cols = ['temperature_celsius', 'feelslike_celsius',
                'air_quality_PM2.5', 'air_quality_PM10',
                'humidity', 'wind_kph', 'precip_mm', 'uv_index', 'air_quality_us-epa-index']

for col in numeric_cols:
    if col in df_bangkok.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

# Remove unnecessary columns

```python
df_bangkok.drop(['country','location_name','latitude','longitude','timezone','last_updated_epoch'], axis=1, inplace=True)
```

```python
df_bangkok.to_csv('bangkok_weather.csv')
print("Cleaned data saved to bangkok_weather.csv")
```
