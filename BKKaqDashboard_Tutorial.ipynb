## General Approach, Justification, and Limitations 

### General Approach  

This dashboard uses a **hybrid setup**:  
- **Plotly (Declarative):** User can describe what they want to show like a scatter plot with specific x and y columns without manually drawing each element.  
- **IPyWidgets (Procedural):** User can control widget such as sliders or checkboxes. In order to make Python functions that filter data and update charts when users change inputs.  

### Why Use This Framework  

- **Jupyter-Friendly:** Everything can runs inside Jupyter, making it easy to build, test, and share dashboards without leaving the environment.  
- **Easy to Learn:** No need to learn additional frameworks like Dash or Streamlit. 
- **Similar Matplotlib structure:** When comparing Plotly Express and Matplotlib, both visualization libraries share similar structures, making it easy for me to choose one of them.

### Limitations  

- **Not for Heavy Traffic:** Voila runs one Python kernel per user, so itâ€™s best for demos, not large-scale deployment.  
- **Complex Callbacks:** Too many interdependent widgets can make the logic harder to manage.  
- **Simple Layouts Only:** ipywidgets layouts (HBox, VBox) are simple compared to modern web dashboards.

### Data Preparation Step
#### From raw data
All step of data preparation before putting in the dashboard will be in the file name `PrepData.ipynb`</br>
The raw data that we get from the website provide many countries and cities data therefore we need to filter only the city that we want to focus.</br>
Steps:
1. Load data
2. Check for duplicate and missing values
3. Extract Bangkok data
4. Convert datatype for datetime and set as index to make it easier to plot graph
5. Create new column for data filtering
6. Drop unnecessary column
7. Save data as a new CSV file

````python
# This code is a summary of PrepData.ipynb
import pandas as pd

# Load the full dataset
df = pd.read_csv('GlobalWeatherRepository.csv')

# --- Check for duplicate and missing values ---
# df.duplicated().sum()  # Result: 0
# df.isnull().sum()     # Result: 0 in all columns

# --- Filter for Bangkok ---
df_bangkok = df[df['location_name'] == 'Bangkok'].copy()

# --- Convert Datetime datatype ---
df_bangkok['last_updated'] = pd.to_datetime(df_bangkok['last_updated'])

# --- Set index (Useful for plotting) ---
df_bangkok.set_index('last_updated', inplace=True)

# --- Feature Engineering (Extract new columns for easier filtering) ---
df_bangkok['month'] = df_bangkok.index.month
df_bangkok['day'] = df_bangkok.index.day
df_bangkok['year'] = df_bangkok.index.year
df_bangkok['hour_of_day'] = df_bangkok.index.time
df_bangkok['day_of_week'] = df_bangkok.index.day_name()

# --- Remove unnecessary columns ---
df_bangkok.drop(['country','location_name','latitude','longitude','timezone','last_updated_epoch'], axis=1, inplace=True)

# --- Save Clean Data as CSV file ---
df_bangkok.to_csv('bangkok_weather.csv')
print("Cleaned data saved to bangkok_weather.csv")
df.head()
````

#### From prepared data to dashboard
Every code from this point will be in the dashboard file named `BKKaqDashboard_plotly.ipynb`

````python
# This code is a part of BkkDashboard.ipynb
# Import libraries
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output

#  Loadfull dataset to make dashboard
df = pd.read_csv("bangkok_weather.csv")

# Convert date column to datetime and format
df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')
df['last_updated'] = df['last_updated'].dt.strftime('%Y-%m-%d')
df.head()
````

## Dashboard making steps


### Visualization Technique
Multiple page:
* **Naviagtion buttons:** putting all graph in one page can be overwhelm for the user to get the insight of the data. So, putting graphs into each page as a section help use to get what we try to convey and monitor in each section.

Filter Type:
* **Multiple Selecttion Filter**: allow user to select multiple of datarange filter to compare each item by itself or as a group range
* **Checkboxes**: use as a toggle condition for the graph like changing from celsius to fahrenheit and kilometer to mile

Graph Type:
1.	**Line Charts:** Ideal for showing trends over time.
    * Plot temperature_celsius over time (using the last_updated column) to show daily or monthly temperature fluctuations.
    * Plot Humidity over time (using the last_updated column) to show daily or monthly humidity fluctuations.
    * Plot Wind speed over time (using the last_updated column) to show daily or monthly chenges.
    * Plot air quality metrics like air_quality_PM2.5 and other air pollutions over time to see pollution trends.
2.	**Bar Charts:** Ideal for comparing categorical data.
    * After extract the month from the last_updated column, we can create a bar chart showing the average precip_mm (precipitation) for each month.
3.	**Scatter Plots:** Showing the relationship between two numerical variables.
    * Does temperature affect humidity? Plot temperature_celsius (x-axis) vs. humidity (y-axis).
    * Does amount of cloud affect UV index? Plot Cloud percentage (x-axis) vs. UV index (y-axis).
4.	**Histogram:** To understand the distribution of a single variable.
    * Plot a histogram of air quality index to see its typical range.

Decoration:
* **Set Layout of Dashboard:** Start with header, navigation buttons, filters, and graph. Every elements in the dashboard will show in the horizontal line.
* **Add style to the header and filter:** Make it beautiful and easier to see

````python
def weather_graphs(df_sub, use_f=False, use_mph=False):
    figs = []

    # --- Temperature Plot ---
    temps = df_sub["temperature_celsius"]
    label = "Â°C"
    if use_f:
        temps = df_sub["temperature_fahrenheit"]
        label = "Â°F"

    fig1 = px.line(
        df_sub,
        x="last_updated",
        y=temps,
        title=f"Temperature ({label})",
        labels={"x": "Date", "y": f"Temperature ({label})"},
        width=475, 
        height=400
    )
    fig1.update_traces(line=dict(color="#f39c12"))
    figs.append(fig1)

    # --- Humidity Plot ---
    fig2 = px.line(
        df_sub,
        x="last_updated",
        y="humidity",
        title="Humidity (%)",
        labels={"x": "Date", "y": "Humidity (%)"},
        width=475,
        height=400
    )
    fig2.update_traces(line=dict(color="#16a085"))
    figs.append(fig2)

    # --- Wind Speed Plot ---
    wind_label = "km/h"
    wind_col = "wind_kph"
    if use_mph:
        wind_col = "wind_mph"
        wind_label = "mph"

    fig3 = px.line(
        df_sub,
        x="last_updated",
        y=wind_col,
        title=f"Wind Speed ({wind_label})",
        labels={"x": "Date", "y": f"Wind Speed ({wind_label})"},
        width=475, 
        height=400
    )
    fig3.update_traces(line=dict(color="#2980b9"))
    figs.append(fig3)

    # --- Rainfall Plot (Average Monthly) ---
    monthly = df_sub.groupby("month")["precip_mm"].mean().dropna().reset_index()
    fig4 = px.bar(
        monthly,
        x="month",
        y="precip_mm",
        title="Avg Monthly Rainfall (mm)",
        labels={"month": "Month", "precip_mm": "Rainfall (mm)"},
        width=450,
        height=400
    )
    fig4.update_traces(marker_color="#3498db")
    figs.append(fig4)

    return figs

# Display graph
weather_figs = weather_graphs(df)
for f in weather_figs:
    f.show()
````

#### Air Quality page
This page will have following grpah:
1. Air pollution Graph - show by the air pollution type
2. Air quality US-EPA Index histogram

<img src="img\PollutionPlot.png">
<img src="img\AQIplot.png">

````python
def air_graphs(df_sub, pollution):
    figs = []

    # --- All pollutants line graph ---
    fig1 = go.Figure()
    for p in pollution:
        if p in df_sub.columns:
            fig1.add_trace(go.Scatter(
                x=df_sub["last_updated"],
                y=df_sub[p],
                mode="lines",
                name=p
            ))
    fig1.update_layout(
        title="Air Quality Pollutants",
        xaxis_title="Date",
        yaxis_title="Pollutant Levels"
    )
    figs.append(fig1)

    # --- AQI Histogram ---
    fig2 = px.histogram(
        df_sub,
        x="air_quality_us-epa-index",
        nbins=10,
        title="US EPA AQI Distribution"
    )
    fig2.update_traces(marker_color="#e74c3c")
    figs.append(fig2)

    return figs

# Display graph
pollutants = ["air_quality_PM2.5", "air_quality_PM10",
    "air_quality_Carbon_Monoxide", "air_quality_Ozone",
    "air_quality_Sulphur_dioxide", "air_quality_Nitrogen_dioxide",]
aq_figs = air_graphs(df, pollutants)
for f in aq_figs:
    f.show()
````

#### Correlation page
This page will have scatterplot grpah including:
1. Relationship of temperature and humidity
2. Relationship of cloud percentage and uv 

<img src="img\scatterPlot1.png">
<img src="img\scatterPlot2.png">

````python
def corr_graphs(df_sub):
    figs = []
    # --- Temperature vs Humidity---
    fig1 = px.scatter(
        df_sub,
        x="temperature_celsius",
        y="humidity",
        title="Temperature vs Humidity",
        opacity=0.6
    )
    fig1.update_traces(marker=dict(color="#27ae60"))
    figs.append(fig1)

    # --- Cloud vs UV Index---
    fig2 = px.scatter(
        df_sub,
        x="cloud",
        y="uv_index",
        title="Cloud % vs UV Index",
        opacity=0.6
    )
    fig2.update_traces(marker=dict(color="#8e44ad"))
    figs.append(fig2)

    return figs

# Display
corr_figs = corr_graphs(df)
for f in corr_figs:
    f.show()
````

### Adding Slicer to the dashboard
The problem with this code is that the slicer and the data is not update when there is some change.</br>
We create `dashboard` variable to add space between each element and make sure they show the max size. 

````python
def render_page(df_sub, page):
    # Toggle slicer visibility based on selected page
    pollutant_box.layout.display = "flex" if page == "ðŸŒ« Air Quality" else "none"
    temp_check_box.layout.display = "flex" if page == "ðŸŒ¤ Weather Overview" else "none"
    wind_check_box.layout.display = "flex" if page == "ðŸŒ¤ Weather Overview" else "none"

    with output:
        clear_output(wait=True)

        # Select which graphs to render
        if page == "ðŸŒ¤ Weather Overview":
            figs = weather_graphs(df_sub, temp_checkbox.value, wind_checkbox.value)
        elif page == "ðŸŒ« Air Quality":
            figs = air_graphs(df_sub, list(pollutant_check.value))
        else:
            figs = corr_graphs(df_sub)

        # If no figures, show message
        if not figs:
            display(widgets.HTML("<p style='color:red;'>No data available.</p>"))
            return

        # Display all Plotly figures side by side
        outputs = []
        for fig in figs:
            out = widgets.Output()
            with out:
                display(fig)  # Plotly figures display directly in Jupyter
            outputs.append(out)

        # Arrange figures in a row
        display(
            widgets.HBox(
                outputs,
                layout=widgets.Layout(
                    justify_content="center",
                    gap="25px",
                    width="100%"
                )
            )
        )
filter_row=widgets.HBox(
    [year_box,month_box,day_box,pollutant_box,temp_check_box,wind_check_box],
    layout=widgets.Layout(justify_content="center",gap="20px",padding="10px")
)
centered_nav_container = widgets.HBox(
    [nav],
    layout=widgets.Layout(justify_content='center')
)
dashboard=widgets.VBox([header,centered_nav_container,filter_row,output],
                       layout=widgets.Layout(width="100%",padding="5px"))
display(dashboard)
````

### Update the graph when choosing filter and update page
We create `df_sub` for data in filter that will go through the graph function of each page. The data will change because of the filter so, we need to make another datafrme that will update the information</br>
Note: There will be missing data because the latest data only cover up to November 2025. So, we add the condition to handle missing data.

````python
def update(*_):
    # Filter data by selected year, month, and day
    df_sub = df[
        df["year"].isin(year_check.value) &
        df["month"].isin(month_check.value) &
        df["day_of_week"].isin(day_check.value)
    ]

    # If no data found, show message
    with output:
        clear_output(wait=True)
        if df_sub.empty:
            display(widgets.HTML("<p style='color:red;'>No data for this selection.</p>"))
            return

    # Render the selected page
    render_page(df_sub, nav.value)

for group in [year_check,month_check,day_check,pollutant_check,nav,temp_checkbox,wind_checkbox]:
    group.observe(update,names="value")

dashboard=widgets.VBox([header,centered_nav_container,filter_row,output],
                       layout=widgets.Layout(width="100%",padding="5px"))
display(dashboard)
update()
````
